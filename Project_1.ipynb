```python
nf['release_year'] = pd.to_numeric(nf['release_year'], errors='coerce')
yearly_trends = nf['release_year'].value_counts().sort_index()
print(yearly_trends)
```

```python


plt.plot(yearly_trends.index, yearly_trends.values , color = 'purple' )
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.title('Netflix Releases Over Time')
plt.grid(True)
plt.show()
```
